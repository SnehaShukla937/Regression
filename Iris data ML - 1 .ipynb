## Data Visualization

```python
sns.pairplot(dsns,hue = 'species', height = 3, aspect = 1) # Plot pairwise relationships in a dataset.
```

```python
dsns.hist(edgecolor = 'black' ,linewidth = 1.2, figsize = (12,8)) # x axis - value range, y -axis = count
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.subplot(2,2,1)
sns.violinplot(x = 'species', y = 'sepal_length', data = dsns)
plt.subplot(2,2,2)
sns.violinplot(x = 'species', y = 'sepal_width', data = dsns)
plt.subplot(2,2,3)
sns.violinplot(x = 'species', y = 'petal_length', data = dsns)
plt.subplot(2,2,4)
sns.violinplot(x = 'species', y = 'petal_width', data = dsns)
```
