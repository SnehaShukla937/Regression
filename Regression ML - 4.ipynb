# Multiple Regression:

Simple Linear Regression: 
$y = \beta _{0} + \beta_{1}X$


Multiple Linear Regression : 
$y = \beta _{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + \beta_{3}X_{3} + ......$


**No. of Features (X) is one in Simple Linear Regression and it is more than one in Multiple Linear Regression.**

### Get significance of any feature using $R^{2}$:

```python
df
```

```python
from sklearn.metrics import r2_score

 # calculation using all features
base_model = smf.ols(formula = 'Y ~ CRIM + ZN  + INDUS+ CHAS + NOX + RM  + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT',data = df)
all_features = base_model.fit()
r2_score(Y,all_features.predict(df))
```

```python
# without LSTAT
test1 = smf.ols(formula = 'Y ~ CRIM + ZN  + INDUS+ CHAS + NOX + RM  + AGE + DIS + RAD + TAX + PTRATIO + B ',data = df)
without_lstat = test1.fit()
r2_score(Y,without_lstat.predict(df))
```

```python
# without AGE
test2 = smf.ols(formula = 'Y ~ CRIM + ZN  + INDUS+ CHAS + NOX + RM  + DIS + RAD + TAX + PTRATIO + B +LSTAT ',data = df)
without_age = test2.fit()
r2_score(Y,without_age.predict(df))
```

#### Here , we an see that:
1. r2 value of features without **LSTAT** reduced from r2 value of all features. That means LSTAT is **most significant** 
    feature of model as it has larger affect on model performance.
    
2. r2 value of features without **AGE** is approx same as r2 value of all features. That means AGE is **least significant**   feature of model as it has no or approx zero effect on model performance.

# Regularized Regression

1. Ridge regression
2. LASSO (Least Absolute Shrinkage and Selection Operator)
3. ElasticNet

1.**Ridge** regression addresses some of the problems of **Ordinary Least Squares** by imposing a penalty on the size of the coefficients. The ridge coefficients minimize a penalized residual sum of squares $(\alpha ||w||_2^2)$:

**$\min_{w} || X w - y||_2^2 + \alpha ||w||_2^2$**

 
The complexity parameter $\alpha \geq 0$ controls the amount of shrinkage: the larger the value of $\alpha$ , the greater the amount of shrinkage and thus the coefficients become more robust to collinearity.

Ridge regression is the **L2** penalized model. Add the **squared** sum of weights to the least square cost function.

..............................................................................................................................

2.The **Lasso** is a linear model that estimates sparse coefficients. 
Mathematically, it consists of a linear model with an added regularization term. The objective function to minimize is:
    
$\min_{w} { \frac{1}{2n_{\text{samples}}} ||X w - y||_2 ^ 2 + \alpha ||w||_1}$

The lasso estimate thus solves the minimization of the least-squares penalty  with $\alpha ||w||_1$
 added, where $\alpha$ is a constant and $||w||_1$ is the **L1-norm** of the coefficient vector.
 
...............................................................................................................................

3.**ElasticNet** is a linear regression model trained with **both L1 and L2-norm regularization** of the coefficients. This combination allows for learning a sparse model where few of the weights are non-zero like Lasso, while still maintaining the regularization properties of Ridge. 

Elastic-net is useful when there are multiple features which are correlated with one another. Lasso is likely to pick one of these at random, while elastic-net is likely to pick both.

A practical advantage of trading-off between Lasso and Ridge is that it allows Elastic-Net to inherit some of Ridgeâ€™s stability under rotation.

The objective function to minimize is in this case:

$\min_{w} { \frac{1}{2n_{\text{samples}}} ||X w - y||_2 ^ 2 + \alpha \rho ||w||_1 +
\frac{\alpha(1-\rho)}{2} ||w||_2 ^ 2}$

### Outlier Impact:

```python
# Linear Regression

from sklearn.linear_model import LinearRegression

np.random.seed(42)
rng = np.random.randn(100) * 10
y_gen = 0.5 * rng + 2 * np.random.randn(100)

lr = LinearRegression()
lr.fit(rng.reshape(-1,1),y_gen)
model_predict = lr.predict(rng.reshape(-1,1))

plt.figure(figsize=(10,8))
plt.scatter(rng,y_gen)
plt.plot(rng,model_predict,c = 'red')
print('estimated coeff. is:',lr.coef_)
```

```python
# Linear Regression after adding outliers

y_gen[rng.argmax()] = 200
y_gen[rng.argmin()] = -200

lrx = LinearRegression(normalize = True)
lrx.fit(rng.reshape(-1,1),y_gen)
model_predictx = lrx.predict(rng.reshape(-1,1))

plt.figure(figsize=(10,8))
plt.scatter(rng,y_gen)
plt.plot(rng,model_predictx,c = 'red')
print('estimated coeff.x is:',lrx.coef_)
```

```python
# Ridge regression or added L2 penalty 

from sklearn.linear_model import Ridge

ridge = Ridge(alpha = 0.5,normalize = True)
ridge.fit(rng.reshape(-1,1),y_gen)
ridge_model_predict = ridge.predict(rng.reshape(-1,1))

plt.figure(figsize=(10,8))
plt.scatter(rng,y_gen)
plt.plot(rng,ridge_model_predict,c = 'red')
print('estimated coeff.x is:',ridge.coef_)
```

```python
# Lasso regression or added L1 penalty 

from sklearn.linear_model import Lasso

lasso = Lasso(alpha = 0.4,normalize = True)
lasso.fit(rng.reshape(-1,1),y_gen)
lasso_model_predict = lasso.predict(rng.reshape(-1,1))

plt.figure(figsize=(10,8))
plt.scatter(rng,y_gen)
plt.plot(rng,lasso_model_predict,c = 'red')
print('estimated coeff.x is:',lasso.coef_)
```

```python
#  ElasticNet regression or added L1+L2 penalty 

from sklearn.linear_model import ElasticNet

en = ElasticNet(alpha = 0.02,normalize = True)
en.fit(rng.reshape(-1,1),y_gen)
en_model_predict = en.predict(rng.reshape(-1,1))

plt.figure(figsize=(10,8))
plt.scatter(rng,y_gen)
plt.plot(rng,en_model_predict,c = 'red')
print('estimated coeff.x is:',en.coef_)
```

1. Estimated Coeff.[actual] = 0.47 <== LR
2. Estimated Coeff.[after adding outliers] = 1.5 <== LR with outliers
3. Estimated Coeff.[after adding L2 penalty] = 1.00 <== Ridge
4. Estimated Coeff.[after adding L1 penalty] = 1.06 <== Lasso
5. Estimated Coeff.[after adding L1+L2 penalty] = 0.74 <=== ElasticNet (**very close to actual estimated value**)
